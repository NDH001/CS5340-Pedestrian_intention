```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.preprocessing import normalize
from matplotlib.pyplot import figure
from pathlib import Path
```

Set the initial path to the dataset

```python
path = Path('online_kl_trajectory_pred.ipynb').parent.absolute().parent.absolute()
print(path)
```

read data from the dataset and split different into columns

```python
df = pd.read_csv(f'{path}/data/origin_dataset/ETH/seq_eth/obsmat.txt')

df.columns =['All']

df = df['All'].str.split('  ', expand=True)
```

Some cleaning on the dataset, dropped pos_z and v_z since it is not used

```python
df.columns = ['index','frame_number', 'pedestrian_ID', 'pos_x' ,'pos_z' ,'pos_y', 'v_x', 'v_z', 'v_y']

df.frame_number = df.frame_number.astype('float64')

df.pedestrian_ID = df.pedestrian_ID.astype('float64')

df.pos_x = df.pos_x.astype('float64')

df.pos_y = df.pos_y.astype('float64')

df.v_x = df.v_x.astype('float64')

df.v_y = df.v_y.astype('float64')

df = df.drop(['pos_z','v_z'],1)
print(df.head())
```

plot all the traj in the dataset to get an overview of the data spread

```python
X_observations = []
Y_observations = []
VX = []
VY = []
temp = df.pedestrian_ID.unique()
for i in range (1,len(temp)):
    temp_row = df.where(df.pedestrian_ID == i).dropna()
    X_observations.append(np.asarray(temp_row.pos_x))
    Y_observations.append(np.asarray(temp_row.pos_y))
    VX.append(np.asarray(temp_row.v_x))
    VY.append(np.asarray(temp_row.v_y))
    

plt.figure(figsize=[40,15])
for i in range(len(X_observations)):
    plt.scatter(X_observations[i],Y_observations[i])

print(len(VX))
print(len(VY))
print(len(X_observations))
```

initialize the acceleartion and delta t

```python
#Initial conditions
initial_acc = 0
delta_t = 1
```

initialize the observation errors, this is shared for both the x and y coordinates

```python
#Observation Errors
delta_x_error = 0.5
delta_x_v_error = 1.5
```

initialize the error in the process covariance matrix, this is shared for both the x and y coordiates

```python
#Process errors in process covariance matrix
delta_p_x = 0.8
delta_p_x_v = 0.8
```

```python
#1.State prediction
#X_k_p = AX_r-1 + B_u_k + W_k

def predicted_state (index,updated_state,axis_observations,axis_v):
  # A will be initialized an a identiy matrix with [0,1] as delta_t
  A = np.identity(2)
  A[0,1] = delta_t

  # initialize X_k-1
  if index == 0:
    X = np.array([axis_observations[index],axis_v[index]])

  # else X_k-1 = X_k
  else:
    X = updated_state
    
  B = np.array([0.5*np.square(delta_t),delta_t])

  W = np.array([initial_acc])
  
  predicted_state = np.matmul(A,X) + B*W
  return A,X,predicted_state
```

```python
#2.Process covariance matrix
def PCM(index,p_PCM,pos_delta,pos_delta_v):
    
  #initialize PCM if first loop
  if index ==0:
    PCM = np.array([[np.square(pos_delta),pos_delta*pos_delta_v],[pos_delta*pos_delta_v,np.square(pos_delta_v)]])

  #else PCM_k-1 = PCM_k
  else:
    PCM = p_PCM
    
  # set all values except the diagonals to 0 for simplicity
  PCM[np.where(np.eye(len(PCM))==0)] = 0
  return PCM
```

```python
#3. P_k = AP_k-1 A^T+ Q_k
# Predicted Process covariance matrix

def predicted_PCM(A,PCM):
  P_PCM = np.matmul(A,PCM)
  P_PCM = np.matmul(P_PCM,A.transpose())
  
  # set all values except the diagonals to 0 for simplicity
  P_PCM[np.where(np.eye(len(P_PCM))==0)] = 0
  return P_PCM
```

```python
#4 Kalman Gain K = (P_k_p H^T) / (H P_k_p H^T + R)

def KGain(P_PCM,delta_error,delta_error_v):
  H = np.identity(len(P_PCM))

  #set up the R matrix
  R = np.identity(len(P_PCM))
  R[0,0] = np.square(delta_error)
  R[1,1] = np.square(delta_error_v)
  
  #(P_k_p H^T)
  top = np.matmul(P_PCM,H.transpose())
  #(H P_k_p H^T + R)
  bot = np.matmul(np.matmul(H,P_PCM),H.transpose())+ R
    
  K = top/bot
  # set all values except the diagonals to 0 for simplicity
  K[np.where(np.eye(len(K))==0)] = 0
  return K
```

```python
# 5 new observation Y_k = CY_K_m + Z_k
def new_obs(index,X,axis_observations,axis_v):
  C = np.identity(len(X))

  Y = np.array([axis_observations[index+1],axis_v[index+1]])
    
  Z = 0
  new_Y = np.matmul(C,Y)+Z
  return new_Y
```

```python
# 6 calculating the current state X_k = X_k + K[Y_k - HX_k]
def current_state(K,predicted_state,new_Y):
    
  # kalman gain calculation, the lesser error is in the observation, the higher K is, and \n
  # the more portion of K[Y_k - HX_k_p] is used to update X_k
  new_X = predicted_state + np.matmul(K,new_Y - np.matmul(np.identity(len(K)),predicted_state))
  return new_X
```

```python
# 7 updating the process covariance matrix
def update_PCM(kg,ppcm):
  I = np.identity(len(kg))
  H = np.identity(len(kg))
  new_P_PCM = np.matmul((I - np.matmul(kg,H)),ppcm)
  return new_P_PCM
```

```python
def kalman_traj(rand):
    
    #initialize three arrays for visual representation and debug
    predicted_states = []
    kalman_filter = []
    uncertainty = []
    #append 0.8 which is the initial coordiate error in the process covariance matrix 
    uncertainty.append(0.8)
    
    # append all of the x and y coordinate into a single array
    all_coor = [X_observations[rand],Y_observations[rand]]
    #append all of the x and y velocity into a single array
    all_coor_v = [VX[rand],VY[rand]]

    # loop two times for x and y individually
    for dim in range(2):
        #initialize X_k-1 and P_k-1
        updated_state = 0
        new_P_PCM = 0
        
        for i in range (len(all_coor[dim])-1):
          
          #calculates the predicted state matrix
          A,X,ps = predicted_state(i,updated_state,all_coor[dim],all_coor_v[dim])
        
          #calculates the predicted process covariance matix
          pcm = PCM(i,new_P_PCM,delta_p_x,delta_p_x_v)
          ppcm = predicted_PCM(A,pcm)
        
          
          #calculates the kalman gain
          kg = KGain(ppcm,delta_x_error,delta_x_v_error)
            
          #calculates the new observation matrix
          updated_obs = new_obs(i,X,all_coor[dim],all_coor_v[dim])
        
          #update X_k 
          updated_state = current_state(kg,ps,updated_obs)
          
          #update P_k
          new_P_PCM = update_PCM(kg,ppcm)
            
          #append data to the arrays accordingly for visualization if needed.
          kalman_filter.append(updated_state[0])
          uncertainty.append(pcm[0][0])
          predicted_states.append(ps[0])
            
    
    # seperates the x and y coordinates' data
    x_predicted_states = predicted_states[0:len(all_coor[0])-1]
    x_kalman_filter = kalman_filter[0:len(all_coor[0])-1]
    x_uncertainty = uncertainty[0:len(all_coor[0])-1]


    y_predicted_states = predicted_states[len(all_coor[0])-1:]
    y_kalman_filter = kalman_filter[len(all_coor[0])-1:]
    y_uncertainty = [uncertainty[0]] + uncertainty[len(all_coor[0])-1:]
    
    
    return x_predicted_states,x_kalman_filter,x_uncertainty,y_predicted_states,y_kalman_filter,y_uncertainty
```

```python
# a utility function that shows the uncertainty in the process covariance matrix for each timestamp
def showfig(x_kalman_filter,y_kalman_filter,x_uncertainty):
    figure(figsize=(20, 10))
    a = 0.1
    for i in range(len(x_kalman_filter)):
        a +=0.05
        a = np.clip(a,0,1)
        plt.gca().add_patch(plt.Circle((x_kalman_filter[i],y_kalman_filter[i] ), x_uncertainty[i], color='r',alpha = a))

    plt.scatter(x_kalman_filter,y_kalman_filter)
    plt.plot(x_kalman_filter, y_kalman_filter,label='kalman')
    plt.grid()
    plt.legend()

```

```python
# pick a pedestrant at random and predict his/her trajectory
rand = np.random.randint(len(VX))
print(rand)
_,x_kalman_filter,x_uncertainty,_,y_kalman_filter,y_uncertainty = kalman_traj(rand)
showfig(x_kalman_filter,y_kalman_filter,x_uncertainty)
```
